Showing how graphics can easily be shown:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x_data = np.arange(0.0, 2.0, 0.01)
y_data = 1 + np.sin(2 * np.pi * x_data)
fig, ax = plt.subplots()
ax.plot(x_data, y_data)
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Sin Example')
plt.show()
```
